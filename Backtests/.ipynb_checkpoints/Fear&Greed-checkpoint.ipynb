```python
import requests
import pandas as pd
import numpy as np
```

```python
r=requests.get('https://api.alternative.me/fng/?limit=0')
```

```python
r
```

```python
df = pd.DataFrame(r.json()['data'])
df
```

```python
df.value = df.value.astype(int)
#Changing timestamp to a readable human format
df.timestamp = pd.to_datetime(df.timestamp, unit='s')
df
```

```python
#setting the index to the timestamp column
df.set_index('timestamp', inplace=True)
df
```

```python
#Illustration of Fear and Greed 
df.plot(figsize=(20,10))
```

```python
#Reversing data from earliest to oldest
df=df[::-1]
df
```

```python
#Adding bitcoin price from yfinance (yahoo)
import yfinance as yf
df1 = yf.download('BTC-USD')[['Close']]
df1
```

```python
#Merging two datasets on the index 
df1.index.name = 'timestamp'
merged = df.merge(df1, on='timestamp')
merged
```

```python
df2 = merged
df2['change'] = df2.Close.pct_change()
```

```python
df2
```

```python
#We are only holding bitcopin if the index value > 50 
df2['position'] = np.where(df2.value > 50, 1, 0)
```

```python
#Calculate the return on the bitcoin - strategy clalculation
strategy = df2.change * df2.position
(strategy+1).cumprod().plot(figsize=(20,10))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.plot((strategy+1).cumprod(), label='F&G')
plt.plot((df2.change+1).cumprod(), label='BTC')
plt.legend()
```
